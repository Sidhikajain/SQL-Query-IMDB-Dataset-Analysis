# **Q8: Find the most awarded average rated movie since Jan 1, 2000**

```python
cur.execute("select original_title from movies where release_date >= 2000-01-01 order by vote_average desc limit 1")
most_awarded_avg=cur.fetchall()
print(f"The most awarded average rated movie is {most_awarded_avg[0][0]}")
```

# **1. Name all the directors with the number of movies and revenue where revenue should be taken into account for doing the analysis.**

# **2. Name all the directors with the number of movies and revenue where number of movies should be taken into account for doing the analysis.**

```python
cur.execute("select name,count(title),sum(revenue) from directors join movies on movies.director_id=directors.id group by directors.id order by count(title) desc")
director_movies=cur.fetchall()
director_most_movies=pd.DataFrame(director_movies,columns=["Director Name","Total Movies","Total Revenue"])
director_most_movies.head(10)
```
